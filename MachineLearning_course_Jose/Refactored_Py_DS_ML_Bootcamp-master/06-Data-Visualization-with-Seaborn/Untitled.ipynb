```python
sns.pairplot(iris, hue="species", diag_kind="hist")
```

```python
iris["PL-PW"]=iris["petal_length"]/iris["petal_width"]
iris["PL-SW"]=iris["petal_length"]/iris["sepal_width"]
iris["PL-SL"]=iris["petal_length"]/iris["sepal_length"]
iris["PW-SW"]=iris["petal_length"]/iris["sepal_width"]
iris["PW-SL"]=iris["petal_length"]/iris["sepal_length"]
iris["SL-SW"]=iris["sepal_length"]/iris["sepal_width"]
```

```python
sns.countplot(x="species", data=iris)
```

```python
#sns.boxplot(x="species", y="PL-PW", data=iris)
#sns.boxplot(x="species", y="PL-SW", data=iris)
sns.boxplot(x="species", y="PW-SW", data=iris)
```

```python
sns.boxplot(x="species", y="sepal_length", data=iris)
```

```python
g = sns.FacetGrid(iris,col="species")
g.map(sns.swarmplot, "sepal_length")
```
